```python
import matplotlib.pyplot as plt
#modul za crtanje grafova i slika
```

```python
seq = [5,8,2,6,1,8,2,6,20,5,6]
seq1 = [1,8,3,4,5,8,2,10,4,54,7]
```

```python
plt.plot(seq)
#vrijednosti seq lisdte postaju y-os
#linearni plot
#sam će cca definirat x os
```

```python
plt.scatter(x=seq, y=seq1)

#zahtjeva definiranje osi
#seq gleda kao x-os, fali mu još y-os da bi mogao plotat
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x=seq,y=seq1, c='blue', s=100, marker='2')
plt.title('Moj prvi python plot')
plt.xlabel('x os')
plt.ylabel('y os')
plt.show()





#napravi mi prazni prostor veličine5x5 i koordinatni sustav na njemu na kojem ću moći crtati graf
#na tom koordinatnom sustavu crtam graf raspršenja osi su mu x=seq i y=seq1, boja oznaka je plava, veličina oznaka je 100, a tip markera je 2
#neka se taj graf zove ' moj prvi python plot'
#x os neka se zove 'x os'
#y os neka se zove 'y os'
#neka se prikaze graf ovdje
```
