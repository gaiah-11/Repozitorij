```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
#od iris podataka napravit dataframe
#.head da dobijemo prvih 5 redova u dataframe
#.tail zadnjih  5 redova
iris_df= pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
iris_df['species'] =iris.target
iris_df.head()
```

```python
iris_df['species'].value_counts()
```

```python
#nakon sto napravimo dictionary njega mapiramo primjenjujemo ga umjesto vrijednosti 0 ce staviti setosa, umjesto 1 versicolor...
species_map={0:'setosa', 1:'versicolor', 2:'virginica'}
iris_df['species']=iris_df['species'].map(species_map)
iris_df['species']
```

```python
iris_df.describe()
```

```python
iris_df['species'].value_counts()
```

T- test

```python
# t test
# iz dataframe-a moramo prvo izdvojiti 2 vrste koje cemo usporedivati
#u stupcu species uzmi sve vrijednosti koje imaju ime setosa i versicolor i njihove druge stupce
set_versi = iris_df[iris_df['species'].isin(['setosa', 'versicolor'])]
set_versi

```

```python
#postoji li statisticki znacajna razlika izmedu grupa
#za setosa izdvoji sepal length

grupa1= set_versi[set_versi['species']=='setosa']['sepal length (cm)']
grupa1
```

```python
grupa2= set_versi[set_versi['species']=='versicolor']['sepal length (cm)']
grupa2
```

```python
#t-test jesu li duljine izmedu te dvije vrsti ide ili nisu jednake odnosno postopji statisticki znacajna razlika
#pingouin ima vec ugradene sve c testove, anova, ...
#p <0.05


ttest= pg.ttest(grupa1,grupa2)
print('T-test rezultati (setosa vs versicolor, sepal length):\n',ttest)

#postoji statisticki znacajna razlika izmedu ove dve grupe odnosno razlika nije slucajna
```

```python
sirina = pg.ttest(iris_df[iris_df['species']== 'setosa']['sepal width (cm)'],
               iris_df[iris_df['species']== 'versicolor']['sepal width (cm)'])
sirina
```

```python
#anova

anova = pg.anova(dv='sepal length (cm)', between='species', data=iris_df)
anova

# velika F vrijednost = razlike izmedu grupa su vece nego razlike unutar grupa
# p je veci od 0.05 razlika nije slucajna nego se radi o drugacijim vrstama cvjetova
```

```python
#korelacija

corr = pg.corr(iris_df['sepal length (cm)'], iris_df['petal length (cm)'])
corr

# r piersonov koeficijent korelacije kada je r = 0 nema linearne korelacije
#r blizu 1 pozitivna korelacija linearna
#r blizu -1 negativna linearna korelacija
# sto je veca duzina listova casice to ce bit veca i duzina latica
```
