```python
from sklearn.impute import  SimpleImputer

#impute = popunjavanje praznih celija
#SimplImputer funkcija kojj se zadaje zadatak popunjavanja praznih celija
#missing values prepoznaje kao nan a strategija popunjavanja je median
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer
```

```python
#napravi dataframe gdje je -666 prazno polje
#napravi simpleimputer koji ce prepoznati -666 i napraviti zamjenu sa konstantnom vrijednosti vrjednosti

imputer = SimpleImputer(missing_values=-666.0, strategy='constant', fill_value=7)
imputer
```

```python
x.fillna(-666.0)
```

```python
# First, ensure the missing values in the input data match what the imputer is expecting.
# Replace np.nan in 'Age' and 'Salary' columns with -666.0.
x_for_imputation = x[['Age','Salary']].fillna(-666.0)

# Now, fit the imputer on this prepared data.
imputer.fit(x_for_imputation)

# Finally, transform the data.
x_imputed = imputer.transform(x_for_imputation)
```

```python
x_imp =pd.DataFrame(x_imputed,
                         columns=['Age','Salary'],
                         index=x.index)
x_imp
```
